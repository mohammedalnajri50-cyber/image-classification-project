```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import random

print("تم استيراد المكتبات بنجاح!")
```

```python
# .. تعني "اذهب للمجلد الأعلى"
data_dir = 'c:/Users/G5/Documents/Visual Studio 2015/projects/image-classification-project/data/Animals'

# هذا الكود يتأكد من أن المسار صحيح
if os.path.exists(data_dir):
    print(f"تم العثور على مجلد البيانات في: {data_dir}")
else:
    print("خطأ: لم يتم العثور على مجلد البيانات! تأكد من أن المجلد موجود في المسار الصحيح.")
```

```python
classes = os.listdir(data_dir)
print(f"وجدنا {len(classes)} فئات:")
print(classes)
```

```python
# إعداد لوحة العرض
num_classes = len(classes)
fig, axes = plt.subplots(1, num_classes, figsize=(15, 5))

# المرور على كل فئة وعرض صورة منها
for i, class_name in enumerate(classes):
    class_path = os.path.join(data_dir, class_name)
    image_files = os.listdir(class_path)
    random_image_file = random.choice(image_files)
    image_path = os.path.join(class_path, random_image_file)
    
    # قراءة الصورة وعرضها
    img = mpimg.imread(image_path)
    axes[i].imshow(img)
    axes[i].set_title(class_name)
    axes[i].axis('off')

# إظهار اللوحة النهائية
plt.show()
```

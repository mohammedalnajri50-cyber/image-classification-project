```python
import os
import matplotlib.pyplot as plt

# مسار مجلد البيانات
base_dir = "C:/Users/G5/Documents/Visual Studio 2015/projects/image-classification-project/data/Animals"  # إذا كنت داخل مجلد notebooks

# أسماء الفئات (المجلدات)
categories = os.listdir(base_dir)
print("الفئات:", categories)

# عرض صورة واحدة من كل فئة
for category in categories:
    folder = os.path.join(base_dir, category)
    images = os.listdir(folder)
    print(f"{category}: عدد الصور = {len(images)}")
    img_path = os.path.join(folder, images[0])
    img = plt.imread(img_path)
    plt.imshow(img)
    plt.title(category)
    plt.axis('off')
    plt.show()
```
